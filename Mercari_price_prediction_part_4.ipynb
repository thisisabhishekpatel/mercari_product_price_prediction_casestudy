#### Analyzing Item Description with Eli5

```python
# Definte RMSLE Cross Validation Function
def rmsle_cv(model):
    kf = KFold(shuffle=True, random_state=42).get_n_splits(data['cleaned_item_description'])
    rmse= np.sqrt(-cross_val_score(model, data['cleaned_item_description'], y, scoring="neg_mean_squared_error", cv = kf))
    return(rmse.mean())
```

```python
from sklearn.linear_model import Ridge

vec = CountVectorizer()
clf = Ridge(random_state=42)
pipe = make_pipeline(vec, clf)
pipe.fit(data['cleaned_item_description'], y)

cv_rmsle = rmsle_cv(pipe)

print("The Validation Score is: " + str(cv_rmsle))
```

#### Feature Pre-Processing / Transformation

```python
from sklearn.pipeline import FeatureUnion

default_preprocessor = CountVectorizer().build_preprocessor()

def build_preprocessor(field):
    field_idx = list(data.columns).index(field)
    return lambda x: default_preprocessor(x[field_idx])

vectorizer = FeatureUnion([
    ('name', CountVectorizer(
        ngram_range=(1, 2),
        max_features=50000,
        preprocessor=build_preprocessor('name'))),
    ('category_main', CountVectorizer(
        token_pattern='.+',
        preprocessor=build_preprocessor('category_main'))),
    ('category_sub1', CountVectorizer(
        token_pattern='.+',
        preprocessor=build_preprocessor('category_sub1'))),
    ('category_sub2', CountVectorizer(
        token_pattern='.+',
        preprocessor=build_preprocessor('category_sub2'))),
    ('brand_name', CountVectorizer(
        token_pattern='.+',
        preprocessor=build_preprocessor('brand_name'))),
    ('shipping', CountVectorizer(
        token_pattern='\d+',
        preprocessor=build_preprocessor('shipping'))),
    ('item_condition_id', CountVectorizer(
        token_pattern='\d+',
        preprocessor=build_preprocessor('item_condition_id'))),
    ('cleaned_item_description', TfidfVectorizer(
        ngram_range=(1, 2),
        max_features=55000,
        stop_words='english',
        preprocessor=build_preprocessor('cleaned_item_description'))),
])
```

#### Modeling

```python
# Create Transformed Train Set
data = vectorizer.fit_transform(data.values)
data
```

```python
def get_rmsle(y, pred): return np.sqrt(mean_squared_error(y, pred))
```

#### Ridge Cross Validation

```python
%%time

# Create 3-Fold CV
cv = KFold(n_splits=3, shuffle=True, random_state=42)
for train_ids, valid_ids in cv.split(data):
    # Define LGBM Model
    model_ridge = Ridge(solver = "lsqr", fit_intercept=False, random_state=42)
    
    # Fit LGBM Model
    model_ridge.fit(data[train_ids], y[train_ids])
    
    # Predict & Evaluate Training Score
    y_pred_train = model_ridge.predict(data[train_ids])
    rmsle_train = get_rmsle(y_pred_train, y[train_ids])
    
    # Predict & Evaluate Validation Score
    y_pred_valid = model_ridge.predict(data[valid_ids])
    rmsle_valid = get_rmsle(y_pred_valid, y[valid_ids])
    
    print(f'LGBM Training RMSLE: {rmsle_train:.5f}')
    print(f'LGBM Validation RMSLE: {rmsle_valid:.5f}')
```

#### LASSO Cross Validation

```python
%%time

# Create 3-Fold CV
cv = KFold(n_splits=3, shuffle=True, random_state=42)
for train_ids, valid_ids in cv.split(data):
    # Define LGBM Model
    model_lasso = Lasso(fit_intercept=True, random_state=42)
    
    # Fit LGBM Model
    model_lasso.fit(data[train_ids], y[train_ids])
    
    # Predict & Evaluate Training Score
    y_pred_train = model_lasso.predict(data[train_ids])
    rmsle_train = get_rmsle(y_pred_train, y[train_ids])
    
    # Predict & Evaluate Validation Score
    y_pred_valid = model_lasso.predict(data[valid_ids])
    rmsle_valid = get_rmsle(y_pred_valid, y[valid_ids])
    
    print(f'LGBM Training RMSLE: {rmsle_train:.5f}')
    print(f'LGBM Validation RMSLE: {rmsle_valid:.5f}')
```

#### LGBM Cross Validation

### Ensemble (Ridge + LGBM)

```python
# Train and Test Split
train_X, test_X, train_y, test_y = train_test_split(data, y, test_size=0.2, random_state=144)
```

```python
# Define LGBM Model
model_lgb = lgb.LGBMRegressor(num_leaves=31, n_jobs=-1, learning_rate=0.1, n_estimators=500, random_state=42)

# Fit LGBM Model
model_lgb.fit(train_X, train_y)

# Predict with LGBM Model
lgbm_y_pred = model_lgb.predict(test_X)
```

```python
# Define Ridge Model
model_ridge = Ridge(solver = "lsqr", fit_intercept=False, random_state=42)
    
# Fit Ridge Model
model_ridge.fit(train_X, train_y)
    
# Evaluate Training Score
ridge_y_pred = model_ridge.predict(test_X)
```

```python
ensemble_y_pred = (lgbm_y_pred+ridge_y_pred)/2

ensemble_rmsle = get_rmsle(ensemble_y_pred, test_y)

print(f'Ensemble RMSLE: {ensemble_rmsle:.5f}')
```
